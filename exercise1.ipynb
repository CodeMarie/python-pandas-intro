```python

literacy_df = pd.read_csv('literacy.csv', encoding='ISO-8859-1')

literacy_df.head(20)
```

```python
year_values = literacy_df.groupby(['Entity', 'Year']).mean().sort_values(by= "Year", ascending=False)
year_values
```

```python

literacy_df['Total Literacy'] = literacy_df['Youth literacy rate, population 15-24 years, male (%)'] + literacy_df['Youth literacy rate, population 15-24 years, female (%)']
literacy_df
```

To see the literacy for both males and female a total literacy was created as a new column. The results were grouped by sorted showing the top 60 results to see the best performing answers. There were multiple values which included: Andorra, Uzbekistan, Azerbaijan and Greenland. 

```python
literacy_comparison = literacy_df.groupby(['Entity', 'Year']).mean().sort_values(by= "Total Literacy", ascending=False).dropna()
literacy_comparison.head(60)
```

The total literacy column was created to combine both female and male data to allow efficiency of the results. The worst Entity was found from using the tail to see the bottom entries. The worst was Burkino Faso. 

```python
literacy_comparison.tail(60)
```

```python
year_1980_df = literacy_df[literacy_df['Year']==1980]
year_1980_df.head(60)

```

```python
year_2000_df = literacy_df[literacy_df['Year']==2000]
year_2000_df.head(60)
```

```python
mean_1980 = year_1980_df.groupby('Year').mean()
mean_1980
```

```python
mean_2000 = year_2000_df.groupby('Year').mean()
mean_2000
```

Sadly, for both genders, the mean for these two years shows that literacy has gone down with a lower mean in 2000 when compared to literacy mean in 1980.
